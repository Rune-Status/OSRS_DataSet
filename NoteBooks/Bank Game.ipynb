```python
class Bank_Account: 
    def __init__(self): 
        self.balance=1000
        self.item_list = []
        print("Hello!!! Welcome to the Deposit & Withdrawal Machine\n")
        
    def search_item(item_id,item_list):
        for row in item_list: 
            if row[0] == item_id:
                return row
            else:
                return None
        
    def buy_item(self,item_id,amount):
        
        # user input must be positive
        if amount < 0:
            print("amount must be positive")
            return
        
        # lookup the current price of the item
        item_price = 100
        
        # if we have the funds to buy the item, remove the funds
        if self.balance >= item_price * abs(amount):
            print("Bought: ", amount)
            self.balance -= item_price * abs(amount)
        else:
            print("Insufficent balance")
            return
        
        # if we already own this item, add to amount if not append to our item_list
        item = search_item(item_id,self.item_list)
        if  item == None:
            # we do not own this item
            self.item_list.append([item_id,amount])
        else:
            # we own this item
            index = self.item_list.index(item)
            self.item_list[index] = [item[0],item[1]+amount]
            
    def sell_item(self,item_id,amount):
        # user input must be positive
        if amount < 0:
            print("amount must be positive")
            # abs(amount)
            return
        
        item = search_item(item_id,self.item_list)
        # check if we have the item
        if  item == None:
            print("User does not own this item")
            return
        
        # check if we have less then than or equal of the amount requested to sell
        if item[1] < amount:
            print("User does not own: ", amount)
            return
        
        # remove the item from the item list
        if item[1] == amount:
            index = self.item_list.index(item)
            self.item_list.pop(index)
        else:
            # add the sold items to our balance
            index = self.item_list.index(item)
            self.item_list[index] = [item[0],item[1]-amount]
            
        # lookup the current price of the item
        item_price = 110
        # add the value of the sold item to our balance
        self.balance += item_price * amount
        
    def display_items(self):
        print(self.item_list)
              
    def display(self): 
        print("\n","Net Available Balance=",self.balance) 
        self.display_items()
```

```python
# Driver code 
   
# creating an object of class 
s = Bank_Account() 
s.display()

# Calling functions with that class object 
s.buy_item(1,10)
s.display()

s.sell_item(1,10)
s.display()
```
